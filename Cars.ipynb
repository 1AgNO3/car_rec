```python
#importing the required dependencies 
import numpy as np
import pandas as pd
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#UPLOADING THE CSV FILE USING PANDAS. 
df = pd.read_csv(r"C:\Users\ACER\Anaconda3\envs\Cars\carrs.csv")
```

```python
df.head(3)
```

```python
#RETURNS THE SHAPE OF THE FILE
df.shape
```

```python
columns = ['Make','Model','MSRP']
df[columns].head(5)
```

```python
#USES AN EMPTY LIST AND ADDS REQUIRED DATA AFTER INSTRUCTION
def get_important_features(data):
    important_features = []
    for i in range(0,data.shape[0]):
        important_features.append(data['Make'][i]+''+data['Model'][i]+''+data['Type'][i]+''+data['Origin'][i])
    
    return important_features
       
```

```python
df['important_features'] = get_important_features(df)
df.head(3)
```

```python
#IT CONVERS THE GIVEN DATA IN THE FORM OF MATRIX 
cm = CountVectorizer().fit_transform(df['important_features'])
```

```python
# COSINE SIMILARITY COMPUTES SIMILARITY AS A NORMALIZED DOT PRODUCT OF X AND Y
cs = cosine_similarity(cm)
print(cs)
```
